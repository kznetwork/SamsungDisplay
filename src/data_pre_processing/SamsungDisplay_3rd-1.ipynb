```python
import requests                  # 웹 페이지의 HTML을 가져오는 모듈
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈

# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get('https://pythondojang.bitbucket.io/weather/observation/currentweather.html')
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', { 'class': 'table_develop3' })    # <table class="table_develop3">을 찾음
data = []                            # 데이터를 저장할 리스트 생성
for tr in table.find_all('tr'):      # 모든 <tr> 태그를 찾아서 반복(각 지점의 데이터를 가져옴)
    tds = list(tr.find_all('td'))    # 모든 <td> 태그를 찾아서 리스트로 만듦
                                     # (각 날씨 값을 리스트로 만듦)
    for td in tds:                   # <td> 태그 리스트 반복(각 날씨 값을 가져옴)
        if td.find('a'):             # <td> 안에 <a> 태그가 있으면(지점인지 확인)
            point = td.find('a').text    # <a> 태그 안에서 지점을 가져옴
            temperature = tds[5].text    # <td> 태그 리스트의 여섯 번째(인덱스 5)에서 기온을 가져옴
            humidity = tds[9].text       # <td> 태그 리스트의 열 번째(인덱스 9)에서 습도를 가져옴
            data.append([point, temperature, humidity])    # data 리스트에 지점, 기온, 습도를 추가

print(data)   # data 표시. 주피터 노트북에서는 print를 사용하지 않아도 변수의 값이 표시됨)
```

```python
#1. make_regression을 이용한 선형회귀에 사용할 데이터 셋 만들기
# 라이브러리를 임포트합니다.
from sklearn.datasets import make_regression

# 특성 행렬, 타깃 벡터, 정답 계수를 생성합니다.
features, target, coefficients = make_regression(n_samples = 100,
                                                 n_features = 3,
                                                 n_informative = 3,
                                                 n_targets = 1,
                                                 noise = 0.0,
                                                 coef = True,
                                                 random_state = 1)

# 특성 행렬과 타깃 벡터를 확인합니다.
print('특성 행렬\n', features[:3])
print('타깃 벡터\n', target[:3])
```

```python
# 2. make_classification을 이용한 분류에 필요한 데이터셋 만들기

# 라이브러리를 임포트합니다.
from sklearn.datasets import make_classification

# 특성 행렬과 타깃 벡터를 생성합니다.
features, target = make_classification(n_samples = 100,
                                       n_features = 3,
                                       n_informative = 3,
                                       n_redundant = 0,
                                       n_classes = 2,
                                       weights = [.25, .75],
                                       random_state = 1)

# 특성 행렬과 타깃 벡터를 확인합니다.
print('특성 행렬\n', features[:3])
print('타깃 벡터\n', target[:3])
```

```python
# 4. make_blobs을 이용한 군집 데이터 셋 그림으로 확인

import matplotlib.pyplot as plt

plt.scatter(features[:,0], features[:,1], c=target)
plt.show()
```

```python
# 3. make_blobs을 이용한 군집에 필요한 데이터 셋 만들기

# 라이브러리를 임포트합니다.
from sklearn.datasets import make_blobs

# 특성 행렬과 타깃 벡터를 생성합니다.
features, target = make_blobs(n_samples = 100,
                              n_features = 2,
                              centers = 3,
                              cluster_std = 0.5,
                              shuffle = True,
                              random_state = 1)

# 특성 행렬과 타깃 벡터를 확인합니다.
print('특성 행렬\n', features[:3])
print('타깃 벡터\n', target[:3])
```

```python
memberInfoDic = {}
with open("datasets/회원정보.txt",'r+', encoding='UTF-8-sig') as f:
    memberInfoList = f.readlines()
    print("회원정보출력\n")
    for line in memberInfoList :
        print(line)
        memberList = line.split(',')
        memberInfoDic[memberList[0]] = [memberList[0],memberList[0].replace('\n','')]
    print(memberInfoDic)
    
    while True:
        userSel = input("\n수정하시려면 '수정', 추가하시려면 '추가', 저장하고 끝내시려면 '끝'을 입력하세요. ")
        if userSel=='수정' or userSel=='추가' or userSel=='끝' :
            if userSel == '수정' :
                alterNum = input("수정하실 회원의 번호를 입력하세요.")
                alterSel = input("이름은 '0', 전화번호를 수정하시려면 '1'을 입력하세요. ")
                alterStr = input("수정할 정보를 입력하세요. ")
                memberInfo = memberInfoDic[alterNum]
                memberInfo[int(alterSel)] = alterStr
                memberInfoDic[alterNum] = memberInfo
                
            elif userSel == '추가' :
                addNum = input("추가하실 회원 번호를 입력하세요. ")
                addName = input("추가하실 회원의 이름을 입력하세요.  ")
                addPhone = input("추가하실 회원의 전화번호를 입력하세요. ")
                memberInfoDic[int(addNum)] = [addName,addPhone]
            else :
                f.seek(0)
                for key,value in memberInfoDic.items() :
                    lineStr = str(key) +','+ value[0] +','+ value[1]+'\n'
                    f.write(lineStr)
                break

            print(memberInfoDic)
        
        else :
            continue
```

```python
#!/usr/bin/env python3
import csv
import glob
import os
import sys

first_file = True
for input_file in glob.glob('./datasets/multicsv/sales_*'):
	print(os.path.basename(input_file))
	with open(input_file, 'r', newline='') as csv_in_file:
		with open("./datasets/multicsv/sales_merge1.csv", 'a', newline='') as csv_out_file:
			filereader = csv.reader(csv_in_file)
			filewriter = csv.writer(csv_out_file)
			if first_file:
				for row in filereader:
					filewriter.writerow(row)
				first_file = False
			else:
				header = next(filereader)
				for row in filereader:
					filewriter.writerow(row)
```

```python
import openpyxl
import copy

rfilename = 'datasets/mergedTable.xlsx'
wfilename = 'datasets/mergedTable_after.xlsx'

def copy_value(rng, sheet):
    value = sheet[rng][0][0].value
    for row in sheet[rng]:
        for col in row:
            col.value = value

wb = openpyxl.load_workbook(rfilename)
sheet = wb.active
mcells = copy.copy(sheet.merged_cells.ranges)
for rng in mcells:
    print(type(rng), str(rng))
    rng = str(rng)
    sheet.unmerge_cells(rng)
    copy_value(rng, sheet)
wb.save(wfilename) 
```

## 폴더 만들기

```python
import os #디렉터리 다루는 라이브러리

with open('./datasets/대전.csv', 'r') as f :
    tempList = f.readlines()

#개행문자 제거
regionList = []
for region in tempList :
    region = region.replace('\n', '')
    regionList.append(region)
#print(regionList)

guList = regionList[0].split(',')
#print(guList)

# '구' 값을 키로 하고 동네 리스트를 Value
regionDic = {}
for guName in guList :
    regionDic[guName] = []

for dongInfo in regionList[1:] :
    dongList = dongInfo.split(',')
    for i, guName in enumerate(guList) :
        tempList = []
        tempList.append(dongList[i]) #['가양동']
        #list(dongList[i]) => '['가','양','동']
        regionDic[guName] += tempList

print(regionDic)

curPath = os.getcwd()
regionName = "대전"
os.mkdir(regionName)

for key, value in regionDic.items() :
    #폴더 이동
    os.chdir(curPath + "\\" + regionName)
    os.mkdir(key)
    for dongName in value :
        if dongName != '' :
            os.chdir(curPath + "\\" + regionName + "\\" + key)
            os.mkdir(dongName)
```

```python
# DB 연결, 커서 획득
# DB 생성 (오토 커밋)
conn = sqlite3.connect("datasets/test.db", isolation_level=None)
# 커서 획득
c = conn.cursor()
# 테이블 생성 (데이터 타입은 TEST, NUMERIC, INTEGER, REAL, BLOB 등)
c.execute("CREATE TABLE IF NOT EXISTS table1 \
    (id integer PRIMARY KEY, name text, birthday text)")

# 데이터 삽입 방법 1
c.execute("INSERT INTO table1 \
    VALUES(1, 'LEE', '1987-00-00')")
```

```python
def read_seq(inputfile):
    with open(inputfile, 'r') as f:
        sequence = f.read()
    sequence = sequence.replace(' ', '')
    sequence = sequence.replace('\n', '')
    sequence = sequence.replace('\r', '')
    return sequence
with open('./datasets/NM_207618.2.fna', 'r') as inf:
    data = inf.read().splitlines(True)
with open('./datasets/dna.txt', 'w') as outf:
    outf.writelines(data[1:])
dna = read_seq('./datasets/dna.txt')
print(dna)
```

```python
def convert(seq):
    """DNA 시퀀스를 아미노산 시퀀스로 변환"""
    genetic_code = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
        'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W',
    }
    protein = ""
    if len(seq) % 3 == 0:  # 데이터의 길이가 3의 배수이면 아래를 실행
        for i in range(0, len(seq), 3):
            codon = seq[i : i+3]
            protein += genetic_code[codon]
    return protein
print(convert(dna[20:938]))
```
